```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np

# 设置字体路径和大小
font_path = "FiraCode-Regular.ttf"  # 例如微软雅黑，可以换成任意 TTF 字体文件

# 加载字体
font = ImageFont.truetype(font_path, 16)

for char in "013":
    # 创建图像（灰度模式，白底）
    image = Image.new('L', (16, 16), color=0)
    draw = ImageDraw.Draw(image)

    # 绘制字符，255表示白色
    draw.text((3, -1), char, font=font, fill=255)

    # 转换为 numpy 数组（0-255），再二值化处理
    bitmap = np.array(image)
    binary_bitmap = (bitmap > 128).astype(int)

    # 打印点阵
    for row in binary_bitmap:
        print(''.join(['#' if pixel else '.' for pixel in row]))
    print()

    # 转化为十六进制数
    result = binary_bitmap.reshape(-1)
    ch = 0
    for i, bit in enumerate(result):
        ch = (ch >> 1) + (int(bit) << 7)
        if i % 8 == 7:
            print(hex(ch), end=", ")
            ch = 0
    print()
    del image
```
