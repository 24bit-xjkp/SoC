```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import common

# 设置字体路径和大小
font_path = "FiraCode-Regular.ttf"  # 例如微软雅黑，可以换成任意 TTF 字体文件
output = (common.SoC_dir / "assets" / "font.data").open("wb")

# 加载字体
font = ImageFont.truetype(font_path, 16)

for char in R""" !"#$%&'()*+,-./0123456789:;<=>?@ABCDEFGHIJKLMNOPQRSTUVWXYZ[\]^_`abcdefghijklmnopqrstuvwxyz{|}~""":
    # 创建图像（灰度模式，白底）
    image = Image.new('L', (16, 16), color=0)
    draw = ImageDraw.Draw(image)

    # 绘制字符，255表示白色
    draw.text((3, -1.5), char, font=font, fill=255)

    # 转换为 numpy 数组（0-255），再二值化处理
    bitmap = np.array(image)
    binary_bitmap = (bitmap > 32).astype(np.uint8)

    # 打印点阵
    for row in binary_bitmap:
        print(''.join(['#' if pixel else '.' for pixel in row]))
    print()

    # 写入文件
    binary_bitmap = np.packbits(binary_bitmap, axis=1)
    output.write(binary_bitmap.tobytes())

    print()
    del image

output.close()
```
